```python

# load dataset
col_names = ['Carat', 'Cut', 'Color', 'Clarity', 'Depth', 'Table', 'Price']
data = pd.read_csv("diamondSmall.csv", header=1, names=col_names)
#take a peek to see if this looks right
data
```

```python
missing_data = data.isnull().sum()
missing_data
```

```python
#I want to handle all of the missing numerical data using the SimpleImputer mean. Compared to the missing categorical data I want to 
# handle using most frequent. 
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

#I looked up how to change the objects into floats so I could use handle the numeric numbers using mean
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
data['Carat'] = pd.to_numeric(data['Carat'], errors='coerce')
data['Depth'] = pd.to_numeric(data['Depth'], errors='coerce')
data['Table'] = pd.to_numeric(data['Table'], errors='coerce')
#numeric features
numeric_features = data[['Carat', 'Depth', 'Table', 'Price']]

imr = SimpleImputer(missing_values=np.nan, strategy='mean')
imputed_data = imr.fit_transform(numeric_features)
data_no_missing = pd.DataFrame(imputed_data, columns=numeric_features.columns)

#categoric data handled using most frequent
categoric_data = data[['Cut', 'Color', 'Clarity']]
imr2 = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputed_data2 = imr2.fit_transform(categoric_data)
data_no_missing_categorical = pd.DataFrame(imputed_data2, columns=categoric_data.columns)

#Putting both of the data together
data_no_missing_concat = pd.concat([data_no_missing_categorical, data_no_missing], axis=1)

data_no_missing_concat
```

```python
#Handling the categorical columns ordinal features : 'Cut'
cut_mapping = {'Premium' : 5, 'Very Good' : 4, 'Ideal' : 3, 'Good' : 2, 'Fair' : 1}
data_no_missing_concat['Cut'] = data_no_missing_concat['Cut'].map(cut_mapping)
data_no_missing_concat
```

```python
#Handling the categorical columns ordinal features : 'Clarity'
clarity_mapping = {'IF' : 7, 'VVS1' : 6, 'VVS2' : 5, 'VS1' : 4, 'VS2' : 3, 'SI1' : 2, 'SI2' : 1}
data_no_missing_concat['Clarity'] = data_no_missing_concat['Clarity'].map(clarity_mapping)
data_no_missing_concat
```

```python
#Handling nominal features
categorical_columns = ['Color']
for column in categorical_columns:
    dummies = pd.get_dummies(data_no_missing_concat[column])
    data_no_missing_concat = pd.concat([ dummies, data_no_missing_concat ], axis = 1)
    data_no_missing_concat = data_no_missing_concat.drop(columns=column)
data_no_missing_concat
                                       
    
```

```python
# I replaced all of the ordinal categorical values : 'Cut' and 'Clarity' by creating a map to convert the string 
# values into integers. I did this because they ordinal features can be sorted and ordered. I ordered 'Cut' from 
# best to worst and I sorted 'Clarity' from highest quality to lowest quality. I then handled the nominal values : 'Color'.
# I did this by using pandas get_dummies method. 
```
